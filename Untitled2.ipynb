The correlation coefficient ranges from â€“1 to 1. When it is close to 1, it means

tends to go up when the median income goes up. When the coefficient is close to
â€“1, it means that there is a strong negative correlation; you can see a small
negative correlation between the latitude and the median house value (i.e., prices
have a slight tendency to go down when you go north). Finally, coefficients close
to zero mean that there is no linear correlation.
